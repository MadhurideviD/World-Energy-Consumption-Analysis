```python
df.shape
```

```python
df = df.dropna(subset=['population'])
```

```python
df['gdp'] = df['gdp'].fillna(df['gdp'].median())
```

```python
# Count NaNs per column
nan_counts = df.isna().sum()
nan_counts = nan_counts[nan_counts > 0].sort_values(ascending=False)
print(nan_counts)
```

Checking correct data types

```python
df['year'] = df['year'].astype(int)
df['country'] = df['country'].astype(str)
df['population'] = df['population'].astype(float)
df['gdp'] = df['gdp'].astype(float)
```

Checking for duplicates

```python
df = df.drop_duplicates()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

Deriving usefull columns

```python
# Fossil fuel total (coal + oil + gas)
df['fossil_consumption'] = df['coal_consumption'] + df['oil_consumption'] + df['gas_consumption']

# Energy type shares
df['fossil_share'] = df['fossil_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['renewable_share'] = df['renewables_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['nuclear_share'] = df['nuclear_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['biofuel_share'] = df['biofuel_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['hydro_share'] = df['hydro_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['solar_share'] = df['solar_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)
df['wind_share'] = df['wind_consumption'] / df['primary_energy_consumption'].replace(0, np.nan)

# Per capita metrics
df['energy_per_capita'] = df['primary_energy_consumption'] / df['population'].replace(0, np.nan)
df['renewables_per_capita'] = df['renewables_consumption'] / df['population'].replace(0, np.nan)
df['fossil_per_capita'] = df['fossil_consumption'] / df['population'].replace(0, np.nan)
```

```python
print(df.isna().sum()[df.isna().sum() > 0])
```
